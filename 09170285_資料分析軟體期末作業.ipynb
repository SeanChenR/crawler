```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from tkinter import *
from bs4 import BeautifulSoup
from bokeh.plotting import figure , output_file , show
from bokeh.models import CategoricalColorMapper

res1 = requests.get('https://www.espn.com/nba/standings')
soup1 = BeautifulSoup(res1.text , 'html.parser')
team_name = soup1.select('abbr')
abb1 = str(team_name[0:31]).split('>')
abb2 = str(abb1[1:61:2]).replace('</abbr','')
abb3 = str(abb2).replace('[','').replace(']','').replace(',','').replace("'","").split(' ')
each_total = []
#球隊薪資總和
for k in abb3:
    res = requests.get('https://www.espn.com/nba/team/roster/_/name/' + k)
    soup = BeautifulSoup(res.text , 'html.parser')
    salary = soup.select('div.inline')
    All = salary[7:len(salary)+1:8]
        
    x = str(All[0:len(All)+1]).replace('<div class="inline">$','').replace('</div>','').replace('<div class="inline">','').replace(',','').replace('[','').replace(']','').split(' ')
    
    SUM = []
    for i in x:
        if i != '--':
            SUM.append(int(i))
    S = sum(SUM)
    each_total.append(S)
print(each_total)
```

```python
#球隊區分成東區西區
east = abb3[0:15]
west = abb3[15:30]
print(east)
print(west)
area = []
for s in abb3:
    if s in east:
        area.append('east')
    else:
        area.append('west')
```

```python
#球隊勝率
data = soup1.select('.stat-cell')
win_pct1 = str(data[2:len(data):13]).replace('<span class="stat-cell">','').replace('</span>','').replace('[','').replace(']','').replace(',','').replace("'","").split(' ')
win_pct2 = []
for q in win_pct1:
    f = '0' + q
    win_pct2.append(float(f))
print(win_pct2)
```

```python
#球隊場均得分
ppg = str(data[8:len(data):13]).replace('<span class="stat-cell">','').replace('</span>','').replace('[','').replace(']','').replace(',','').replace("'","").split(' ')
point_per_game = []
for t in ppg:
    point_per_game.append(float(t))
print(point_per_game)
```

```python
#球隊場均失分
opg = str(data[9:len(data):13]).replace('<span class="stat-cell">','').replace('</span>','').replace('[','').replace(']','').replace(',','').replace("'","").split(' ')
opponent_point_per_game = []
for s in opg:
    opponent_point_per_game.append(float(s))
print(opponent_point_per_game)
```
