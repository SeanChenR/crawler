```python
import requests

from bs4 import BeautifulSoup

url = "https://www.books.com.tw/web/sys_bbotm/books/020901/?o=1&v=1&page=5"

for times in range(5):
    res = requests.get(url)
    soup = BeautifulSoup(res.text , 'html.parser')
    articles = soup.select('h4 a')
    paging = soup.select('.page a')
    
    last_url = paging[0]['href']
    
    for each_title in articles:
        print(each_title.text , each_title['href'])
        
    url = last_url
```
